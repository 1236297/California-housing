```python

import numpy as np
import torch as torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader , TensorDataset
import matplotlib.pyplot as plt
import spacy
import os

```

```python
import pandas as pd

Data_path_train = '/content/sample_data/california_housing_train.csv'
Data_path_test = '/content/sample_data/california_housing_test.csv'

# Load the data into Pandas DataFrames
Data_path_train = pd.read_csv(Data_path_train)
Data_path_test = pd.read_csv(Data_path_test)

# Check for missing values
print(Data_path_train.isnull().sum())
print(Data_path_test.isnull().sum())
```

```python
Data_path_train.fillna(Data_path_train.median(), inplace=True)
Data_path_test.fillna(Data_path_test.median(), inplace=True)
```

```python
Data_path_train = pd.get.Dummies(Data_path_train, columns = ['ocean_proximity'] )
Data_path_test = pd.get.Dummies(Data_path_test, columns = ['ocean_proximity'])
Data_path_train, Data_path_test = Data_path_train.align(Data_path_test, join='left', axis=1 )
Data_path_train.fillna(Data_path_train.median(), inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
X.train = Data_path_train.drop('target', axis=1)
y.train = Data_path_train['target']
X.test = Data_path_test.drop('target', axis=1)
y.test = Data_path_test['target']
```

```python
scaler = StandardScaler()
X.train = scaler.fit_transform(X.train)
X.test = scaler.transform(X.test)
```

```python
X_train_tensor = torch.tensor(X.train, dtype=torch.float32)
y_train_tensor = torch.tensor(y.train.values, dtype=torch.float32)
X_val_tensor = torch.tensor(X.test, dtype=torch.float32)
y_val_tensor = torch.tensor(y.test.values, dtype=torch.float32)
```
